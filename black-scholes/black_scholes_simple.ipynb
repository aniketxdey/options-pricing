# Black-Scholes Model Calculator

A calculator for the Black-Scholes Model Calculator using Python.

## Background

The Black-Scholes Model is commonly used in finance for pricing options, specifically European call-options, which are purchased on pre-determined dates. The Black-Scholes model: calculates the theoretical value of an option based on five variables:

*   Underlying Price (S)
*   Strike Price (K)
*   Time To Expiration (T)
*   Risk-Free Interest Rate (r)
*   Volatility (σ)



## 1. Define Necessary Libraries

```python
import math
from scipy.stats import norm
```

## 2. Define Variables

```python
# Define the variables - Edit to mirror values in the data
S = 42 # Underlying Price
K = 40 # Strike Price
T = 0.5 # Time to Expiration
r = 0.1 # Risk-Free Interest Rate
vol = 0.2 # Volatitlity (σ)
```

## 3. Calculating d1

The formula for d1 in the Black-Scholes model is:
Screenshot 2024-07-18 at 5.04.30 PM.png


```python
d1 = (math.log(S/K) + (r + 0.5 * vol**2)*T ) / (vol * math.sqrt(T))
```

## 4. Calculating d2

The formula for d2 in the Black-Scholes model is:

Screenshot 2024-07-18 at 5.06.26 PM.png

```python
d2 = d1 - (vol * math.sqrt(T))
```

## 5. Calculate Call Option Price

The call option price (C) in the Black-Scholes Model is calculated with the formula:

Screenshot 2024-07-18 at 5.07.54 PM.png

```python
C = S * norm.cdf(d1) - K * math.exp(-r * T) * norm.cdf(d2)
```

## 6. Calculate Put Option

The put option price (P) is calculated as:

Screenshot 2024-07-18 at 5.08.57 PM.png

```python
  P = K * math.exp(-r * T) * norm.cdf(-d2) - S * norm.cdf(-d1)
```

```python
print('The value of d1 is: ', round(d1, 4))
print('The value of d2 is: ', round(d2, 4))
print('The price of the call option is: $', round(C, 2))
print('The price of the put option is: $', round(P, 2))
```
